```python
%load_ext autoreload
%autoreload 2

import numpy as np
import ehtplot
from matplotlib import pyplot as plt

import metronization as mt
```

```python
N = 256

x = ((np.arange(N) + 0.5) / N - 0.5) * 320
y = ((np.arange(N) + 0.5) / N - 0.5) * 320
X, Y = np.meshgrid(x, y, indexing='ij')

def Gaussian(x0, y0, sigma=60):    
    return np.exp(-((X-x0)**2 + (Y-y0)**2) / sigma**2)

img = Gaussian(0,0,60) - Gaussian(5,3,50)

plt.imshow(img.T, origin='lower', cmap='afmhot_10us')
```

```python
mt.metronize(img, 32, plot=True, threshold=0.2)
mt.toposign (img, 32, plot=True, threshold=0.2)
```

```python
mt.metronize(img, 32, plot=True, threshold=0.3)
mt.toposign (img, 32, plot=True, threshold=0.3, hbirth=10)
```
