```python
%load_ext autoreload
%autoreload 2

import numpy as np
import ehtplot
from matplotlib import pyplot as plt

import metroization as mt
```

```python
# create grids in image and visibility domains

N = 256

x = ((np.arange(N) + 0.5) / N - 0.5) * 320
y = ((np.arange(N) + 0.5) / N - 0.5) * 320
X, Y = np.meshgrid(x, y)

def Gaussian(x0, y0, sigma=60):    
    return np.exp(-((X-x0)**2 + (Y-y0)**2) / sigma**2)

img = Gaussian(0,0,60) - Gaussian(5,0,50)

plt.imshow(img, origin='lower', cmap='afmhot_10us')
```
